```python
# ĞŸĞ¾Ğ´ĞºĞ»ÑÑ‡ĞµĞ½Ğ¸Ğµ Ğ±Ğ¸Ğ±Ğ»Ğ¸Ğ¾Ñ‚ĞµĞºĞ¸ Ğ´Ğ»Ñ Ğ´Ğ¾ÑÑ‚ÑƒĞ¿Ğ° Ğº Ñ„Ğ°Ğ¹Ğ»Ğ°Ğ¼
from google.colab import files
# Ğ˜Ğ¼Ğ¿Ğ¾Ñ€Ñ‚Ğ¸Ñ€ÑƒĞµĞ¼ Ğ´Ğ»Ñ Ñ€Ğ°Ğ±Ğ¾Ñ‚Ñ‹ Ñ Ñ„Ğ°Ğ¹Ğ»Ğ°Ğ¼Ğ¸
import shutil
import os
# Ğ”Ğ»Ñ Ñ€Ğ°Ğ±Ğ¾Ñ‚Ñ‹ Ñ ÑĞ°Ğ±Ğ¿Ñ€Ğ¾Ñ†ĞµÑÑĞ°Ğ¼Ğ¸ (Ğ±Ñ‹ÑÑ‚Ñ€ĞµĞµ, Ñ‡ĞµĞ¼ Ñ‡ĞµÑ€ĞµĞ· !)
import subprocess
# Ğ”Ğ»Ñ ÑƒÑĞºĞ¾Ñ€ĞµĞ½Ğ¸Ñ Ğ¿Ñ€Ğ¾Ñ†ĞµÑÑĞ°
import torch

# 1ï¸âƒ£ Ğ—Ğ°Ğ³Ñ€ÑƒĞ·ĞºĞ° Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸
!pip install openai-whisper ffmpeg
```

Ğ—Ğ°Ğ¿ÑƒÑĞº Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸ Ñ‚Ñ€Ğ°Ğ½ÑĞºÑ€Ğ¸Ğ±Ğ°Ñ†Ğ¸Ğ¸ Ğ¿Ğ¾ Ñ„Ğ°Ğ¹Ğ»Ñƒ

```python

# 2ï¸âƒ£ ĞÑ‚ĞºÑ€Ñ‹Ğ²Ğ°ĞµĞ¼ Ğ¾ĞºĞ½Ğ¾ Ğ·Ğ°Ğ³Ñ€ÑƒĞ·ĞºĞ¸
uploaded = files.upload()

# 3ï¸âƒ£ Ğ‘ĞµÑ€Ñ‘Ğ¼ Ğ¸Ğ¼Ñ Ğ·Ğ°Ğ³Ñ€ÑƒĞ¶ĞµĞ½Ğ½Ğ¾Ğ³Ğ¾ Ñ„Ğ°Ğ¹Ğ»Ğ°
source_video_file = list(uploaded.keys())[0]

# 4ï¸âƒ£ ĞŸÑ€Ğ¾Ğ²ĞµÑ€ÑĞµĞ¼, ĞµÑĞ»Ğ¸ Ñ„Ğ°Ğ¹Ğ» ÑƒĞ¶Ğµ Ğ² Ñ„Ğ¾Ñ€Ğ¼Ğ°Ñ‚Ğµ mp3, Ñ‚Ğ¾ Ğ¿Ñ€Ğ¾Ğ¿ÑƒÑĞºĞ°ĞµĞ¼ ÑˆĞ°Ğ³ ĞºĞ¾Ğ½Ğ²ĞµÑ€Ñ‚Ğ°Ñ†Ğ¸Ğ¸
if source_video_file.lower().endswith(".mp3"):
    shutil.move(source_video_file, "audio.mp3")
    print("âœ… Ğ—Ğ°Ğ³Ñ€ÑƒĞ¶ĞµĞ½ MP3 Ñ„Ğ°Ğ¹Ğ», ĞºĞ¾Ğ½Ğ²ĞµÑ€Ñ‚Ğ°Ñ†Ğ¸Ñ Ğ½Ğµ Ñ‚Ñ€ĞµĞ±ÑƒĞµÑ‚ÑÑ.")
else:
    destination_path = "video.mp4"
    shutil.move(source_video_file, destination_path)
    print(f"âœ… Ğ¤Ğ°Ğ¹Ğ» Ğ·Ğ°Ğ³Ñ€ÑƒĞ¶ĞµĞ½ Ğ¸ ÑĞ¾Ñ…Ñ€Ğ°Ğ½Ñ‘Ğ½ ĞºĞ°Ğº {destination_path}")

    # 5ï¸âƒ£ ĞŸĞµÑ€ĞµĞ´ĞµĞ»Ğ°ĞµĞ¼ Ğ²Ğ¸Ğ´ĞµĞ¾ Ñ„Ğ¾Ñ€Ğ¼Ğ°Ñ‚ Ğº Ğ°ÑƒĞ´Ğ¸Ğ¾
    subprocess.run(["ffmpeg", "-i", "video.mp4", "-vn", "-acodec", "mp3", "audio.mp3", "-loglevel", "error"])

# 6ï¸âƒ£ ĞŸÑ€Ğ¾Ğ²ĞµÑ€ÑĞµĞ¼ Ğ½Ğ°Ğ»Ğ¸Ñ‡Ğ¸Ğµ GPU Ğ¸ Ğ²Ñ‹Ğ±Ğ¸Ñ€Ğ°ĞµĞ¼ ÑƒÑÑ‚Ñ€Ğ¾Ğ¹ÑÑ‚Ğ²Ğ¾
use_cuda = torch.cuda.is_available()
device_flag = "--device cuda" if use_cuda else "--fp16 False"

# 7ï¸âƒ£ Ğ£Ğ´Ğ°Ğ»ÑĞµĞ¼ ÑÑ‚Ğ°Ñ€Ñ‹Ğ¹ `text_transcription.txt`, ĞµÑĞ»Ğ¸ Ğ¾Ğ½ ĞµÑÑ‚ÑŒ
new_text_file = "text_transcription.txt"
if os.path.exists(new_text_file):
    os.remove(new_text_file)
    print("ğŸ—‘ï¸ Ğ£Ğ´Ğ°Ğ»Ñ‘Ğ½ ÑÑ‚Ğ°Ñ€Ñ‹Ğ¹ Ñ„Ğ°Ğ¹Ğ» text_transcription.txt")

# 8ï¸âƒ£ Ğ—Ğ°Ğ¿ÑƒÑĞº Ñ‚Ñ€Ğ°Ğ½ÑĞºÑ€Ğ¸Ğ±Ğ°Ñ†Ğ¸Ğ¸ Ñ Whisper
whisper_cmd = f"whisper audio.mp3 --model base {device_flag} --output_format txt"
subprocess.run(whisper_cmd.split(), capture_output=True, text=True)

# 9ï¸âƒ£ ĞŸÑ€Ğ¾Ğ²ĞµÑ€ÑĞµĞ¼, Ğ¿ĞµÑ€ĞµĞ¸Ğ¼ĞµĞ½Ğ¾Ğ²Ñ‹Ğ²Ğ°ĞµĞ¼ Ğ¸ ÑĞºĞ°Ñ‡Ğ¸Ğ²Ğ°ĞµĞ¼ Ñ„Ğ°Ğ¹Ğ»
text_file_name = "audio.txt"

if os.path.exists(text_file_name):
    shutil.move(text_file_name, new_text_file)
    print(f"âœ… Ğ¢Ñ€Ğ°Ğ½ÑĞºÑ€Ğ¸Ğ¿Ñ†Ğ¸Ñ Ğ·Ğ°Ğ²ĞµÑ€ÑˆĞµĞ½Ğ° Ğ¸ ÑĞ¾Ñ…Ñ€Ğ°Ğ½ĞµĞ½Ğ° ĞºĞ°Ğº {new_text_file}")
    files.download(new_text_file)
else:
    print("â— ĞÑˆĞ¸Ğ±ĞºĞ°: Ñ„Ğ°Ğ¹Ğ» Ñ‚Ñ€Ğ°Ğ½ÑĞºÑ€Ğ¸Ğ¿Ñ†Ğ¸Ğ¸ Ğ½Ğµ Ğ½Ğ°Ğ¹Ğ´ĞµĞ½.")

# ğŸ”Ÿ Ğ£Ğ´Ğ°Ğ»ĞµĞ½Ğ¸Ğµ Ğ²Ñ€ĞµĞ¼ĞµĞ½Ğ½Ñ‹Ñ… Ñ„Ğ°Ğ¹Ğ»Ğ¾Ğ²
for temp_file in ["audio.mp3", "video.mp4", "audio.json", "audio.srt", "audio.tsv", "audio.vtt"]:
    if os.path.exists(temp_file):
        os.remove(temp_file)
        print(f"ğŸ—‘ï¸ Ğ£Ğ´Ğ°Ğ»Ñ‘Ğ½ Ğ²Ñ€ĞµĞ¼ĞµĞ½Ğ½Ñ‹Ğ¹ Ñ„Ğ°Ğ¹Ğ»: {temp_file}")
```
