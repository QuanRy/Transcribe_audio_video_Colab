# **Транскрибация - это просто!**

❗Загрузка модели (ЕДИНОЖДЫ на рабочее время)❗

```python
# Подключение библиотеки для доступа к файлам
from google.colab import files
# Импортируем для работы с файлами
import shutil
import os
# Для работы с сабпроцессами (быстрее, чем через !)
import subprocess
# Для ускорения процесса
import torch

# 1️⃣ Загрузка модели
!pip install openai-whisper ffmpeg
```

Запуск модели транскрибации по файлу

```python

# 2️⃣ Открываем окно загрузки
uploaded = files.upload()

# 3️⃣ Берём имя загруженного файла
source_video_file = list(uploaded.keys())[0]

# 4️⃣ Проверяем, если файл уже в формате mp3, то пропускаем шаг конвертации
if source_video_file.lower().endswith(".mp3"):
    shutil.move(source_video_file, "audio.mp3")
    print("✅ Загружен MP3 файл, конвертация не требуется.")
else:
    destination_path = "video.mp4"
    shutil.move(source_video_file, destination_path)
    print(f"✅ Файл загружен и сохранён как {destination_path}")

    # 5️⃣ Переделаем видео формат к аудио
    subprocess.run(["ffmpeg", "-i", "video.mp4", "-vn", "-acodec", "mp3", "audio.mp3", "-loglevel", "error"])

# 6️⃣ Проверяем наличие GPU и выбираем устройство
use_cuda = torch.cuda.is_available()
device_flag = "--device cuda" if use_cuda else "--fp16 False"

# 7️⃣ Удаляем старый `text_transcription.txt`, если он есть
new_text_file = "text_transcription.txt"
if os.path.exists(new_text_file):
    os.remove(new_text_file)
    print("🗑️ Удалён старый файл text_transcription.txt")

# 8️⃣ Запуск транскрибации с Whisper
whisper_cmd = f"whisper audio.mp3 --model base {device_flag} --output_format txt"
subprocess.run(whisper_cmd.split(), capture_output=True, text=True)

# 9️⃣ Проверяем, переименовываем и скачиваем файл
text_file_name = "audio.txt"

if os.path.exists(text_file_name):
    shutil.move(text_file_name, new_text_file)
    print(f"✅ Транскрипция завершена и сохранена как {new_text_file}")
    files.download(new_text_file)
else:
    print("❗ Ошибка: файл транскрипции не найден.")

# 🔟 Удаление временных файлов
for temp_file in ["audio.mp3", "video.mp4", "audio.json", "audio.srt", "audio.tsv", "audio.vtt"]:
    if os.path.exists(temp_file):
        os.remove(temp_file)
        print(f"🗑️ Удалён временный файл: {temp_file}")
```
